```python
# from the baseline 2.2 run
import numpy as np
import matplotlib.pylab as plt
from rubin_scheduler.utils import ddf_locations

%matplotlib inline
from make_ddf_survey import optimize_ddf_times, generate_ddf_scheduled_obs

from scipy.stats import binned_statistic
```

```python
ddfs = ddf_locations()

ddf_data = np.load('/Users/yoachim/rubin_sim_data/scheduler/ddf_grid.npz')
ddf_grid = ddf_data['ddf_grid'].copy()

ddf_name = 'XMM_LSS'
```

```python
ack = ddf_grid["sun_alt"][0:-1] * ddf_grid["sun_alt"][1:]
night = np.zeros(ddf_grid.size, dtype=int)
night[np.where((ddf_grid["sun_alt"][1:] >= 0) & (ack < 0))] += 1
night = np.cumsum(night)
```

```python
bins = np.arange(ddf_grid["mjd"][0], ddf_grid["mjd"][-1], 1)

mags = ddf_grid[ddf_name+'_m5_g'] + 0
mags[np.isinf(mags)] = np.nan

bs, be, bn = best_mag_night = binned_statistic(ddf_grid["mjd"], mags, statistic=np.nanmax, bins=bins)
```

```python
fig, ax = plt.subplots()

x = bins - np.min(bins)
ax.plot(x[1:], bs)
ax.set_xlim([0,365*2])
ax.set_xlabel('Night')
ax.set_ylabel('Best 5-sigma g depth (mags)')
ax.set_title(ddf_name)
```
